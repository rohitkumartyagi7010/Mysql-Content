# Step-by-Step Guide to Creating and Using Functions in MySQL

Functions in MySQL are used to:

1.Perform operations and calculations.

2.Encapsulate and reuse complex logic.

3.Simplify queries and make them more readable.

MySQL provides many built-in functions for various tasks. For example:

a. Mathematical Functions: ABS(), ROUND(), FLOOR(), CEIL(), etc.

b. String Functions: CONCAT(), SUBSTRING(), LENGTH(), REPLACE(), etc.

c. Date and Time Functions: NOW(), CURDATE(), DATE_FORMAT(), etc.

c. Aggregate Functions: SUM(), AVG(), COUNT(), MAX(), MIN(), etc.
