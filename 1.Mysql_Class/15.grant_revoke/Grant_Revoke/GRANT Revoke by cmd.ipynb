# Grant and Revoke Commands

# 1. Create a New User

As an admin, you can create a new user by running the following command:

```python
#syntax
CREATE USER 'newuser'@'localhost' IDENTIFIED BY 'password';
#Replace 'newuser' with the desired username and 'password' with a secure password for the new user.
```

```python
# create user sachin with password 
CREATE USER 'sachin'@'localhost' IDENTIFIED BY 'admin';
```

Grant the SELECT permission on a specific database (e.g., database_name) to the newly created user:

# 2. Grant Permission

```python
# syntax
# database_name (choose database you want to select)
# * denotes all tables you can select any particular table if you want (example : sakila.actor)
GRANT SELECT ON database_name.* TO 'newuser'@'localhost';
```

```python
# granting select pemission on all tables of sakila database
# * denotes all tables you can select any particular table to provide access
GRANT SELECT ON sakila.* TO 'sachin'@'localhost';
```

```python
# providing permission on sakila database and actor table
GRANT SELECT ON sakila.actor TO 'sachin'@'localhost';
```

# 3. Apply Changes (Flush Privileges)

After creating the user and granting permissions, you should apply the changes by flushing the privileges:

# 4. Grant More Privileges (e.g., INSERT, UPDATE)

If you want to grant more permissions like INSERT or UPDATE, you can run:

```python
GRANT INSERT, UPDATE ON sakila.* TO 'sachin'@'localhost';
```

This will allow the user to perform both INSERT and UPDATE operations on the tables within the database.

# 5. Revoke Permissions

```python
REVOKE INSERT, UPDATE ON database_name.* FROM 'user_name'@'localhost';
```

```python
REVOKE INSERT, UPDATE ON sakila.* FROM 'user_name'@'localhost';
```

# 6. Remove All Privileges

```python
REVOKE ALL PRIVILEGES ON database_name.* FROM 'user_name'@'localhost';
```

```python
REVOKE ALL PRIVILEGES ON sakila.* FROM 'sachin'@'localhost';
```

# 7. Delete the User (Optional)

If you want to delete the user from MySQL, use the following command:

```python
DROP USER 'user_name'@'localhost';
```

```python
DROP USER 'sachin'@'localhost';
```
